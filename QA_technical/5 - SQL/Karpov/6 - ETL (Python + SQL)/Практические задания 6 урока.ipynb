**–°—Ä–µ–¥–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –Ω–∏–∂–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤—ã–±–µ—Ä–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π:**

![](https://storage.yandexcloud.net/klms-public/production/learning-content/11/77/598/1724/7851/image.png)



```python
query = """
SELECT room_type,
       toFloat32OrNull(replaceRegexpAll(price, '[$,]', '')) as price,
       toFloat32OrNull(review_scores_rating) as review_scores_rating
FROM ldn_listings
WHERE review_scores_rating is not NULL
LIMIT 1000
"""
df = ph.read_clickhouse(query, connection=connection)
df.head()
```

```python
mean_price = df.groupby('room_type').price.mean()
mean_price
```

```python
mean_rating = df.groupby('room_type').review_scores_rating.mean()
mean_rating
```

```python
mean_df = df.groupby('room_type', as_index=False).agg({'price' : 'mean', 'review_scores_rating' : 'mean'})
mean_df
```

```python
sns.scatterplot(data=mean_df, x="price", y="review_scores_rating", hue="room_type")
sns.despine()
```

```python
query = """
SELECT DISTINCT host_id as uniq_host,
       host_verifications
FROM ldn_listings
WHERE experiences_offered != 'none'
"""
df_verif = ph.read_clickhouse(query, connection=connection)
df_verif.head()
```

```python
df_verif.host_verifications = df_verif.host_verifications.apply(lambda x: x.strip(']').strip('[').replace('\'', '').split(', '))
df_verif = df_verif.explode('host_verifications')
df_verif.head()
```

**–ö–∞–∫–æ–π –≥—Ä–∞—Ñ–∏–∫ –ø–æ–ª—É—á–∏–ª—Å—è?**

![](https://storage.yandexcloud.net/klms-public/production/learning-content/11/77/598/1724/7853/image2.png)

```python
query  = """
SELECT
    COUNT(experiences_offered) as experience_count,
    experiences_offered,
    neighbourhood_cleansed
FROM
    ldn_listings
WHERE
    experiences_offered !='none'
GROUP BY 
    neighbourhood_cleansed,
    experiences_offered
LIMIT 100
    """
exp_df = ph.read_clickhouse(query, connection=connection)
exp_df.head()
```

```python
exp_df_type = exp_df.groupby(['neighbourhood_cleansed', 'experiences_offered'], as_index=False) \
                    .agg({'experience_count' : 'sum'}) \
                    .sort_values('experience_count', ascending=False)
```

```python
pivot_exp = exp_df_type.pivot(index='neighbourhood_cleansed', columns='experiences_offered', values='experience_count') \
                       .fillna(0).head(100)
```

```python
sns.set(
    font_scale =2,
    style      ="whitegrid",
    rc         ={'figure.figsize':(20,7)}
       )
cmap=sns.cubehelix_palette(as_cmap=True)
sns.heatmap(pivot_exp, cmap=cmap)
sns.despine()
```

* –ù–∞ –≤—Ç–æ—Ä–æ–º ‚Äì –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (np.log())

```python
query  = """
SELECT
    room_type,
    toFloat32OrNull(replaceRegexpAll(price, '[$,]', '')) AS price
FROM
    ldn_listings
WHERE
    experiences_offered !='none'
    """
rt_df = ph.read_clickhouse(query, connection=connection)
rt_df.head()
```

```python
sns.distplot(rt_df.query("room_type == 'Private room'").price, kde=False, label='Private room')
sns.distplot(rt_df.query("room_type == 'Entire home/apt'").price, kde=False, label='Entire home/apt')
sns.distplot(rt_df.query("room_type == 'Hotel room'").price, kde=False, label='Hotel room')
sns.distplot(rt_df.query("room_type == 'Shared room'").price, kde=False, label='Shared room')
plt.legend()
plt.show()
```

Hints: line plot; —É–¥–∞–ª–∏—Ç–µ –ø–æ—Å–ª–µ–¥–Ω–∏–µ 6 —Å–∏–º–≤–æ–ª–æ–≤ —É –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ –¥–∞—Ç—ã, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å –≥–æ–¥, –Ω–∞–ø—Ä–∏–º–µ—Ä, —Å –ø–æ–º–æ—â—å—é –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è lambda-—Ñ—É–Ω–∫—Ü–∏–∏.

**–°—Ä–µ–¥–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –Ω–∏–∂–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤—ã–±–µ—Ä–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π.**

![](https://storage.yandexcloud.net/klms-public/production/learning-content/11/77/598/1724/7855/image.png)

```python
query  = """
SELECT  
    room_type,
    toFloat32OrNull(replaceRegexpAll(price, '[$,]', '')) AS price_corr,
    first_review
FROM ldn_listings 
WHERE 
    first_review > '2010-01-01'
LIMIT 1000
    """
review_df = ph.read_clickhouse(query, connection=connection)
review_df.head()
```

```python
review_df.first_review = review_df.first_review.apply(lambda x: x[:4])
review_df.head()
```

```python
avg_prices = review_df.groupby(['room_type', 'first_review'], as_index=False).agg({'price_corr' : 'mean'})
```

```python
sns.lineplot(x='first_review', y='price_corr', hue='room_type', data=avg_prices)
sns.despine()
```
